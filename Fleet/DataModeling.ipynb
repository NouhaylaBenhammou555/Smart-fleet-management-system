```python
import folium
import pandas as pd
from folium import FeatureGroup
from folium.plugins import MarkerCluster

#Create the base Map
m = folium.Map(location=[37.090240,-95.712891], tiles='OpenStreetMap', zoom_start=5)

#Read the data
df = pd.read_csv("MyDataset.csv")

#Create the markers
markerCluster = MarkerCluster().add_to(m)
for i, row in df.iterrows():
    lat  = df.at[i,'latitude']
    lng = df.at[i,'longitude']
    camion = df.at[i,'Type Camion']
    popup = df.at[i,'Type Camion'] 

    if camion == 'Camion frigorifique':
        color = 'blue'
    elif camion == 'Grand Camion':
        color = 'red'
    else:
        color = 'purple'

    folium.Marker(location=[lat, lng],popup=camion, icon=folium.Icon(color=color)).add_to(markerCluster)
m
 
```
